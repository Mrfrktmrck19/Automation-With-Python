*Then, we can move on to generate multiple Cover Letters.* <br>
We took the data from excel. We are gonna fill the blank these data and generate CLs more than one.

```python
data.columns
```

```python
for index,row in data.iterrows():
  myContext = {"company_name":row["Company"],
               "hr_names":row["HR_Name"],
               "hr_email":row["Email"],
               "company_address":row["Address"],
               "job":row["Job"],
               "title":row["Title"]
               }
  context.update(myContext)
  myDocument.render(context)
  myDocument.save(f"newGenerated{index}.docx")
myContext
```
